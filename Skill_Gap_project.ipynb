```python
data['Skills'] = data['Skills'].astype(str)

# convert to lowercase
data['Skills'] = data['Skills'].str.lower()

# split using ;
data['Skills'] = data['Skills'].apply(lambda x: x.split(';'))

# remove spaces
data['Skills'] = data['Skills'].apply(lambda x: [i.strip() for i in x])

# remove empty and nan
data['Skills'] = data['Skills'].apply(lambda x: [i for i in x if i != '' and i != 'nan'])
```

```python
from collections import Counter

# count all skills
all_skills = [skill for sublist in data['Skills'] for skill in sublist]
skill_counts = Counter(all_skills)

# keep only skills appearing at least 5 times
common_skills = {skill for skill, count in skill_counts.items() if count >= 5}

# filter skills
data['Skills'] = data['Skills'].apply(lambda x: [s for s in x if s in common_skills])
```

```python
data = data[data['Skills'].apply(len) > 0]
data = data.reset_index(drop=True)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
X = mlb.fit_transform(data['Skills'])
y = data['Title']
```

```python
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# model
model = KNeighborsClassifier(n_neighbors=3)

# train
model.fit(X_train, y_train)

print("Model trained successfully ✅")
```

```python
print(X.shape)
```

```python
user_skills = ["python", "sql"]

user_input = mlb.transform([user_skills])

prediction = model.predict(user_input)

print("Predicted Role:", prediction[0])
```

```python
# ===== USER INPUT =====
user_input_skills = input("Enter your skills (comma separated): ").lower().split(",")

# clean spaces
user_input_skills = [i.strip() for i in user_input_skills]

# transform
user_input = mlb.transform([user_input_skills])

# prediction
prediction = model.predict(user_input)

print("\nPredicted Role:", prediction[0])


# ===== SKILL GAP =====
role_skills = data.groupby('Title')['Skills'].apply(lambda x: set(sum(x, [])))

required = role_skills[prediction[0]]
missing = required - set(user_input_skills)

print("Missing Skills:", missing)


# ===== MATCH % =====
match = len(set(user_input_skills) & required)
total = len(required)

percentage = (match / total) * 100

print("Match Percentage:", round(percentage, 2), "%")
```
